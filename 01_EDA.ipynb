```python
# ============================================================
# Cella 1: Import delle librerie
# Spiegazione: importiamo tutti gli strumenti che useremo.
# Dare un "alias" (pd, np, plt, sns) è convenzione standard.
# ============================================================

import pandas as pd           # per lavorare con tabelle (DataFrame)
import numpy as np            # per calcoli numerici
import matplotlib.pyplot as plt  # per creare grafici
import seaborn as sns          # grafici statistici più belli
import warnings
warnings.filterwarnings('ignore')   # nascondi avvisi non critici

# Impostazioni grafici
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
sns.set_theme(style='whitegrid', palette='husl')

print("Librerie importate con successo!")
```

```python
# ============================================================
# Cella 2: Caricamento del dataset
# pd.read_csv() legge un file CSV e crea un DataFrame.
# Un DataFrame è come una tabella Excel in Python.
# ============================================================

# Prova prima con il file principale, poi con quello expanded se esiste
import os

# Adatta questo nome al file effettivamente scaricato
possible_files = [
    'data/StudentPerformanceFactors.csv'
]

df = None
for path in possible_files:
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f"Dataset caricato da: {path}")
        break

if df is None:
    print("ATTENZIONE: nessun file trovato. Controlla la cartella data/")
    print("File presenti:", os.listdir('data'))
else:
    print(f"\nDimensioni dataset: {df.shape[0]} righe × {df.shape[1]} colonne")
```

```python
# ============================================================
# Cella 3: Prima occhiata al dataset
# .head() mostra le prime 5 righe (come vedere l'intestazione di Excel)
# ============================================================

print("=== PRIME 5 RIGHE ===")
df.head()  # In Jupyter, l'ultima riga viene visualizzata come tabella
```

```python
# ============================================================
# Cella 4: Informazioni strutturali
# .info() dice: quante righe, che tipo è ogni colonna,
# e quanti valori non-nulli (ovvero: ci sono dati mancanti?)
# ============================================================

print("=== INFORMAZIONI SUL DATASET ===")
df.info()

print("\n=== NOMI DELLE COLONNE ===")
print(df.columns.tolist())

print("\n=== TIPI DI DATI ===")
print(df.dtypes)
```

```python
# ============================================================
# Cella 5: Statistiche descrittive
# .describe() calcola: media, deviazione standard, min, max,
# quartili (25%, 50%=mediana, 75%) per le colonne numeriche.
# ============================================================

print("=== STATISTICHE COLONNE NUMERICHE ===")
df.describe().round(2)
```

```python
# ============================================================
# Cella 6: Analisi dei valori mancanti (NaN = Not a Number)
# In ogni dataset reale ci sono dati mancanti. Dobbiamo
# sapere quanti ce ne sono e in quali colonne.
# ============================================================

missing = df.isnull().sum()
missing_perc = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Valori mancanti': missing,
    'Percentuale %': missing_perc
}).sort_values('Valori mancanti', ascending=False)

print("=== VALORI MANCANTI PER COLONNA ===")
print(missing_df[missing_df['Valori mancanti'] > 0])

if missing.sum() == 0:
    print("Nessun valore mancante! Il dataset è completo.")
else:
    print(f"\nTotale valori mancanti: {missing.sum()}")
```

```python
# ============================================================
# Cella 7: Pulizia del dataset
# Strategia: eliminiamo TUTTE le righe che contengono
# almeno un valore mancante (NaN), in qualsiasi colonna.
# ============================================================

df_clean = df.copy()  # lavoriamo sempre su una copia

righe_prima = len(df_clean)

df_clean = df_clean.dropna()  # elimina ogni riga con almeno un NaN

righe_dopo = len(df_clean)
righe_eliminate = righe_prima - righe_dopo

print(f"Righe prima della pulizia:    {righe_prima}")
print(f"Righe eliminate (con NaN):    {righe_eliminate} ({righe_eliminate/righe_prima*100:.1f}%)")
print(f"Righe dopo la pulizia:        {righe_dopo}")
print(f"Valori mancanti rimasti:      {df_clean.isnull().sum().sum()}")
```
