```python
# ============================================================
# Cella 1: Import delle librerie
# Spiegazione: importiamo tutti gli strumenti che useremo.
# Dare un "alias" (pd, np, plt, sns) è convenzione standard.
# ============================================================

import pandas as pd           # per lavorare con tabelle (DataFrame)
import numpy as np            # per calcoli numerici
import matplotlib.pyplot as plt  # per creare grafici
import seaborn as sns          # grafici statistici più belli
import warnings
warnings.filterwarnings('ignore')   # nascondi avvisi non critici

# Impostazioni grafici
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
sns.set_theme(style='whitegrid', palette='husl')

print("Librerie importate con successo!")
```

```python
# ============================================================
# Cella 2: Caricamento del dataset
# pd.read_csv() legge un file CSV e crea un DataFrame.
# Un DataFrame è come una tabella Excel in Python.
# ============================================================

# Prova prima con il file principale, poi con quello expanded se esiste
import os
os.makedirs('grafici', exist_ok=True)

# Adatta questo nome al file effettivamente scaricato
possible_files = [
    'data/StudentPerformanceFactors.csv'
]

df = None
for path in possible_files:
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f"Dataset caricato da: {path}")
        break

if df is None:
    print("ATTENZIONE: nessun file trovato. Controlla la cartella data/")
    print("File presenti:", os.listdir('data'))
else:
    print(f"\nDimensioni dataset: {df.shape[0]} righe × {df.shape[1]} colonne")
```

```python
# ============================================================
# Cella 3: Prima occhiata al dataset
# .head() mostra le prime 5 righe (come vedere l'intestazione di Excel)
# ============================================================

print("=== PRIME 5 RIGHE ===")
df.head()  # In Jupyter, l'ultima riga viene visualizzata come tabella
```

```python
# ============================================================
# Cella 4: Informazioni strutturali
# .info() dice: quante righe, che tipo è ogni colonna,
# e quanti valori non-nulli (ovvero: ci sono dati mancanti?)
# ============================================================

print("=== INFORMAZIONI SUL DATASET ===")
df.info()

print("\n=== NOMI DELLE COLONNE ===")
print(df.columns.tolist())

print("\n=== TIPI DI DATI ===")
print(df.dtypes)
```

```python
# ============================================================
# Cella 5: Statistiche descrittive
# .describe() calcola: media, deviazione standard, min, max,
# quartili (25%, 50%=mediana, 75%) per le colonne numeriche.
# ============================================================

print("=== STATISTICHE COLONNE NUMERICHE ===")
df.describe().round(2)
```

```python
# ============================================================
# Cella 6: Analisi dei valori mancanti (NaN = Not a Number)
# In ogni dataset reale ci sono dati mancanti. Dobbiamo
# sapere quanti ce ne sono e in quali colonne.
# ============================================================

missing = df.isnull().sum()
missing_perc = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Valori mancanti': missing,
    'Percentuale %': missing_perc
}).sort_values('Valori mancanti', ascending=False)

print("=== VALORI MANCANTI PER COLONNA ===")
print(missing_df[missing_df['Valori mancanti'] > 0])

if missing.sum() == 0:
    print("Nessun valore mancante! Il dataset è completo.")
else:
    print(f"\nTotale valori mancanti: {missing.sum()}")
```

```python
# ============================================================
# Cella 7: Pulizia del dataset
# Strategia: eliminiamo TUTTE le righe che contengono
# almeno un valore mancante (NaN), in qualsiasi colonna.
# ============================================================

df_clean = df.copy()  # lavoriamo sempre su una copia

righe_prima = len(df_clean)

df_clean = df_clean.dropna()  # elimina ogni riga con almeno un NaN

righe_dopo = len(df_clean)
righe_eliminate = righe_prima - righe_dopo

print(f"Righe prima della pulizia:    {righe_prima}")
print(f"Righe eliminate (con NaN):    {righe_eliminate} ({righe_eliminate/righe_prima*100:.1f}%)")
print(f"Righe dopo la pulizia:        {righe_dopo}")
print(f"Valori mancanti rimasti:      {df_clean.isnull().sum().sum()}")
```

```python
# ============================================================
# Cella 8: Analisi variabili categoriche
# .select_dtypes(include=['object']) seleziona SOLO le colonne
# che contengono testo (stringhe). Le numeriche vengono ignorate.
# value_counts() conta quante volte appare ogni valore unico.
# ============================================================

cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
print(f"Colonne categoriche trovate ({len(cat_cols)}): {cat_cols}\n")

for col in cat_cols:
    print(f"{'='*40}")
    print(f"Colonna: {col}")
    print(f"Valori unici: {df_clean[col].nunique()}")
    conteggio = df_clean[col].value_counts()
    percentuali = (df_clean[col].value_counts(normalize=True) * 100).round(1)
    tabella = pd.DataFrame({'Conteggio': conteggio, 'Percentuale %': percentuali})
    print(tabella)
    print()
```

```python
# ============================================================
# Cella 9: Bar chart per ogni variabile categorica
# subplots() crea una griglia di grafici.
# n_rows e n_cols vengono calcolati automaticamente in base
# al numero di colonne categoriche trovate.
# ============================================================

cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
n_cols_plot = 2
n_rows_plot = (len(cat_cols) + 1) // 2

fig, axes = plt.subplots(n_rows_plot, n_cols_plot, figsize=(14, n_rows_plot * 4))
axes = axes.flatten() if n_rows_plot > 1 else axes

for i, col in enumerate(cat_cols):
    counts = df_clean[col].value_counts()
    palette = sns.color_palette('husl', len(counts))
    axes[i].bar(counts.index, counts.values, color=palette, edgecolor='white', linewidth=0.5)
    axes[i].set_title(f"Distribuzione: {col}", fontsize=13, fontweight='bold')
    axes[i].set_xlabel(col, fontsize=11)
    axes[i].set_ylabel('Conteggio', fontsize=11)
    axes[i].tick_params(axis='x', rotation=30)
    # aggiunge il numero sopra ogni barra
    for j, v in enumerate(counts.values):
        axes[i].text(j, v + 5, f'{v}', ha='center', fontsize=10)

for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.savefig('grafici/distribuzione_categoriche.png', dpi=150, bbox_inches='tight')
plt.show()
print("Grafico salvato.")
```

```python
# ============================================================
# Cella 10: Istogrammi con KDE per le colonne numeriche
# kde=True aggiunge la curva di densità stimata.
# axvline() disegna una linea verticale alla posizione indicata.
# ============================================================

# Seleziona colonne numeriche (escludendo eventuali ID o colonne derivate)
num_cols = df_clean.select_dtypes(include=['number']).columns.tolist()
print(f"Colonne numeriche trovate: {num_cols}")

n_plot = len(num_cols)
fig, axes = plt.subplots((n_plot+1)//2, 2, figsize=(14, ((n_plot+1)//2) * 4))
axes = axes.flatten()
colors = ['#4f46e5', '#06b6d4', '#10b981', '#f59e0b', '#ef4444', '#8b5cf6']

for i, col in enumerate(num_cols):
    media = df_clean[col].mean()
    mediana = df_clean[col].median()
    sns.histplot(df_clean[col], kde=True, ax=axes[i],
                 color=colors[i % len(colors)], bins=30, alpha=0.7)
    axes[i].axvline(media, color='red', linestyle='--', linewidth=1.5,
                    label=f'Media: {media:.1f}')
    axes[i].axvline(mediana, color='orange', linestyle='--', linewidth=1.5,
                    label=f'Mediana: {mediana:.1f}')
    axes[i].set_title(f'Distribuzione: {col}', fontweight='bold')
    axes[i].set_xlabel(col); axes[i].set_ylabel('Frequenza')
    axes[i].legend(fontsize=10)

for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.savefig('grafici/distribuzione_numeriche.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# Cella 11: Boxplot e calcolo outlier con metodo IQR
# IQR = Q3 - Q1 (range del 50% centrale dei dati)
# Outlier = valori < Q1 - 1.5*IQR  oppure  > Q3 + 1.5*IQR
# ============================================================

num_cols = df_clean.select_dtypes(include=['number']).columns.tolist()

fig, ax = plt.subplots(figsize=(12, 5))
bp = df_clean[num_cols].boxplot(
    ax=ax, patch_artist=True,
    boxprops=dict(facecolor='#6366f1', alpha=0.7, color='#312e81'),
    medianprops=dict(color='white', linewidth=2.5),
    whiskerprops=dict(color='#94a3b8'),
    capprops=dict(color='#94a3b8'),
    flierprops=dict(marker='o', color='#ef4444', alpha=0.5, markersize=4)
)
ax.set_title('Boxplot variabili numeriche – punti rossi = outlier', fontweight='bold')
ax.set_ylabel('Valore')
ax.tick_params(axis='x', rotation=20)
plt.tight_layout()
plt.savefig('grafici/boxplot_outlier.png', dpi=150, bbox_inches='tight')
plt.show()

# Report testuale degli outlier
print("\n=== REPORT OUTLIER PER COLONNA ===")
print(f"{'Colonna':<30} {'Q1':>6} {'Q3':>6} {'IQR':>6} {'Outlier':>8} {'%':>6}")
print("-"*65)
for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_out = len(df_clean[(df_clean[col] < lower) | (df_clean[col] > upper)])
    perc = n_out / len(df_clean) * 100
    print(f"{col:<30} {Q1:>6.1f} {Q3:>6.1f} {IQR:>6.1f} {n_out:>8} {perc:>5.1f}%")
```

```python
# ============================================================
# Cella 12: Heatmap della matrice di correlazione
# .corr() calcola la correlazione di Pearson tra tutte
# le coppie di colonne numeriche.
# mask=np.triu(...) nasconde il triangolo superiore perché
# è identico a quello inferiore (la matrice è simmetrica).
# ============================================================

num_cols = df_clean.select_dtypes(include=['number']).columns.tolist()
corr_matrix = df_clean[num_cols].corr()

fig, ax = plt.subplots(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
    mask=mask, ax=ax, vmin=-1, vmax=1,
    linewidths=0.5, linecolor='#1a1d27',
    annot_kws={'size': 10, 'weight': 'bold'},
    cbar_kws={'shrink': 0.8}
)
ax.set_title('Matrice di correlazione (Pearson) – solo triangolo inferiore',
             fontweight='bold', pad=15)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('grafici/heatmap_correlazione.png', dpi=150, bbox_inches='tight')
plt.show()

# Stampa coppie con correlazione alta (>0.5 o <-0.5)
print("\n=== COPPIE CON CORRELAZIONE SIGNIFICATIVA (|r| > 0.5) ===")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        val = corr_matrix.iloc[i, j]
        if abs(val) > 0.5:
            c1 = corr_matrix.columns[i]
            c2 = corr_matrix.columns[j]
            direzione = "positiva" if val > 0 else "negativa"
            print(f"  {c1} ↔ {c2}: r = {val:.3f} (correlazione {direzione})")
```

```python
# ============================================================
# Cella 13: Confronto Exam_Score per genere
# violinplot: la larghezza di ogni "violino" in un punto
# indica quanti studenti hanno quel punteggio.
# inner='quartile' mostra le linee Q1, mediana, Q3 dentro.
# ============================================================

# Trova la colonna genere (prova nomi comuni)
gender_col = None
for c in df_clean.columns:
    if 'gender' in c.lower() or 'sex' in c.lower():
        gender_col = c; break

score_col = 'Exam_Score' if 'Exam_Score' in df_clean.columns else \
            [c for c in df_clean.columns if 'score' in c.lower() or 'Score' in c][0]

if gender_col:
    fig, axes = plt.subplots(1, 2, figsize=(13, 6))

    # Violin plot
    sns.violinplot(data=df_clean, x=gender_col, y=score_col,
                   ax=axes[0], palette='Set2', inner='quartile')
    axes[0].set_title(f'Distribuzione {score_col} per {gender_col}', fontweight='bold')

    # Boxplot a confronto
    sns.boxplot(data=df_clean, x=gender_col, y=score_col,
                ax=axes[1], palette='Set2')
    axes[1].set_title(f'Boxplot {score_col} per {gender_col}', fontweight='bold')

    plt.tight_layout()
    plt.savefig('grafici/punteggio_per_genere.png', dpi=150, bbox_inches='tight')
    plt.show()

    print(f"\n=== STATISTICHE {score_col.upper()} PER {gender_col.upper()} ===")
    print(df_clean.groupby(gender_col)[score_col].agg(['mean','median','std','min','max']).round(2))
else:
    print("Colonna genere non trovata. Controlla df_clean.columns")
```

```python
# ============================================================
# Cella 14: Effetto delle sessioni di tutoraggio sul punteggio
# Tutoring_Sessions = numero di sessioni di tutoraggio seguite
# ============================================================

prep_col = 'Tutoring_Sessions'
score_col = 'Exam_Score'

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

sns.boxplot(data=df_clean, x=prep_col, y=score_col, ax=axes[0], palette='coolwarm')
axes[0].set_title(f'{score_col} vs {prep_col}', fontweight='bold')
axes[0].set_xlabel('Numero di sessioni di tutoraggio')
axes[0].set_ylabel('Punteggio esame')

# Istogrammi sovrapposti per ogni numero di sessioni
for gruppo in sorted(df_clean[prep_col].unique()):
    subset = df_clean[df_clean[prep_col] == gruppo][score_col]
    sns.histplot(subset, ax=axes[1], kde=True, label=f'{gruppo} sessioni', alpha=0.4)
axes[1].set_title('Distribuzione punteggio per n° sessioni tutoraggio', fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.savefig('grafici/effetto_tutoraggio.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n=== MEDIA PUNTEGGIO PER NUMERO DI SESSIONI ===")
print(df_clean.groupby(prep_col)[score_col].agg(['mean','median','count']).round(2))
```

```python
# ============================================================
# Cella 15: Scatter plot tra coppie di variabili numeriche
# np.polyfit() calcola la retta di regressione lineare.
# np.poly1d() costruisce la funzione della retta.
# alpha=0.4 rende i punti semitrasparenti (si vede meglio
# dove si sovrappongono molti studenti).
# ============================================================

num_cols = df_clean.select_dtypes(include=['number']).columns.tolist()

# Prendi al massimo 3 coppie significative
pairs = [(num_cols[i], num_cols[j])
         for i in range(len(num_cols))
         for j in range(i+1, len(num_cols))][:3]

fig, axes = plt.subplots(1, len(pairs), figsize=(6*len(pairs), 5))
if len(pairs) == 1: axes = [axes]

for idx, (c1, c2) in enumerate(pairs):
    axes[idx].scatter(df_clean[c1], df_clean[c2],
                      alpha=0.35, color='#6366f1', s=20, edgecolors='none')
    # Linea di tendenza
    z = np.polyfit(df_clean[c1], df_clean[c2], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df_clean[c1].min(), df_clean[c1].max(), 200)
    axes[idx].plot(x_line, p(x_line), 'r-', linewidth=2, label='Trend lineare')
    r = df_clean[[c1, c2]].corr().iloc[0, 1]
    axes[idx].set_title(f'{c1} vs {c2}\nr = {r:.3f}', fontweight='bold')
    axes[idx].set_xlabel(c1); axes[idx].set_ylabel(c2)
    axes[idx].legend()

plt.tight_layout()
plt.savefig('grafici/scatter_plot.png', dpi=150, bbox_inches='tight')
plt.show()
```
