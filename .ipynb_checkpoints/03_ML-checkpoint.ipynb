```python
# ============================================================
# Cella ML1: Setup del notebook ML
# Carichiamo i dati già pronti dal preprocessing.
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# Modelli
from sklearn.linear_model  import LogisticRegression
from sklearn.tree          import DecisionTreeClassifier, plot_tree
from sklearn.ensemble      import RandomForestClassifier

# Metriche
from sklearn.metrics import (accuracy_score, classification_report,
                               confusion_matrix, ConfusionMatrixDisplay,
                               roc_curve, auc)
from sklearn.model_selection import cross_val_score

os.makedirs('grafici', exist_ok=True)
sns.set_theme(style='whitegrid')

# Carica i dati preprocessati
X_train = pd.read_csv('data/X_train_scaled.csv')   # versione scalata per modelli lineari
X_test  = pd.read_csv('data/X_test_scaled.csv')
y_train = pd.read_csv('data/y_train.csv').squeeze()  # squeeze() converte DataFrame → Series
y_test  = pd.read_csv('data/y_test.csv').squeeze()

feature_cols = X_train.columns.tolist()

print("=== DATI CARICATI ===")
print(f"X_train: {X_train.shape}  |  y_train: {y_train.shape}")
print(f"X_test:  {X_test.shape}   |  y_test:  {y_test.shape}")
print(f"Feature: {feature_cols}")
```

```python
# ============================================================
# Cella ML2: Addestramento e confronto di 3 modelli
# ============================================================


modelli = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Decision Tree':      DecisionTreeClassifier(max_depth=5, random_state=42),
    'Random Forest':      RandomForestClassifier(n_estimators=100, random_state=42)
}

risultati = {}

print(f"{'='*72}")
print(f"{'Modello':<25} {'Acc. Train':>10} {'Acc. Test':>10} {'Gap':>8} {'Overfitting?':>13} {'Score':>8}")
print(f"{'='*72}")

for nome, modello in modelli.items():
    modello.fit(X_train, y_train)

    acc_train = accuracy_score(y_train, modello.predict(X_train))
    acc_test  = accuracy_score(y_test,  modello.predict(X_test))
    y_pred    = modello.predict(X_test)
    gap       = acc_train - acc_test

    # Penalizza i modelli con overfitting: se il gap supera il 5%,
    # sottraiamo metà del gap in eccesso dallo score finale.
    # Es: acc_test=0.88, gap=0.10 → penalità=0.025 → score=0.855
    # Es: acc_test=0.86, gap=0.02 → penalità=0    → score=0.860
    penalita  = (gap - 0.05) * 0.5 if gap > 0.05 else 0
    score     = acc_test - penalita
    overfit   = "SI ⚠" if gap > 0.05 else "no ✓"

    risultati[nome] = {
        'modello': modello, 'y_pred': y_pred,
        'acc_train': acc_train, 'acc_test': acc_test,
        'gap': gap, 'score': score
    }
    print(f"{nome:<25} {acc_train:>10.3f} {acc_test:>10.3f} {gap:>8.3f} {overfit:>13} {score:>8.3f}")

print(f"{'='*72}")

# Il miglior modello è quello con lo score più alto,
# che bilancia accuracy sul test set e assenza di overfitting.
best_name = max(risultati, key=lambda k: risultati[k]['score'])

print(f"\nMiglior modello (accuracy + stabilità): {best_name}")
print(f"  Accuracy test:  {risultati[best_name]['acc_test']:.3f} ({risultati[best_name]['acc_test']*100:.1f}%)")
print(f"  Gap train/test: {risultati[best_name]['gap']:.3f} (overfitting {'rilevato ⚠' if risultati[best_name]['gap'] > 0.05 else 'assente ✓'})")
print(f"  Score corretto: {risultati[best_name]['score']:.3f}")
```

```python
# ============================================================
# Cella ML3: Analisi approfondita del miglior modello
# ============================================================

best = risultati[best_name]
y_pred_best = best['y_pred']

print(f"=== REPORT COMPLETO: {best_name} ===")
print(f"\nAccuracy complessiva: {best['acc_test']:.3f} ({best['acc_test']*100:.1f}%)")
print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred_best,
      target_names=['Low Performance (0)', 'High Performance (1)']))

# Confusion Matrix
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

cm = confusion_matrix(y_test, y_pred_best)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=['Low Perf.', 'High Perf.'])
disp.plot(ax=axes[0], cmap='Blues', colorbar=False)
axes[0].set_title(f'Confusion Matrix – {best_name}', fontweight='bold')

# Confusion Matrix normalizzata (in percentuale)
cm_norm = confusion_matrix(y_test, y_pred_best, normalize='true')
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm_norm,
                                display_labels=['Low Perf.', 'High Perf.'])
disp2.plot(ax=axes[1], cmap='Greens', colorbar=False)
axes[1].set_title('Confusion Matrix normalizzata (% per classe reale)', fontweight='bold')

plt.tight_layout()
plt.savefig('grafici/confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"""
Interpretazione:
  Veri Negativi  (TN): {cm[0][0]:4d} → Low Perf. predetti Low Perf.   ✓
  Falsi Positivi (FP): {cm[0][1]:4d} → Low Perf. predetti High Perf.  ✗
  Falsi Negativi (FN): {cm[1][0]:4d} → High Perf. predetti Low Perf.  ✗
  Veri Positivi  (TP): {cm[1][1]:4d} → High Perf. predetti High Perf. ✓
""")
```

```python
# ============================================================
# Cella ML4: Cross-Validation a 5 fold
# ============================================================
#
# Il singolo train/test split dipende da COME i dati sono
# stati divisi. Con una divisione sfortunata potresti avere
# risultati troppo ottimistici o pessimistici.
#
# La k-fold cross-validation risolve questo:
# 1. Divide il training set in 5 "fold" (pezzi)
# 2. Addestra il modello 5 volte, ogni volta usando:
#    - 4 fold come training
#    - 1 fold come validazione (ruota ogni volta)
# 3. Fa la media delle 5 accuracy ottenute
# Risultato: stima più stabile e affidabile.
# ============================================================

print("=== CROSS-VALIDATION (5 fold sul training set) ===")
print(f"\n{'Modello':<25} {'Media':>8} {'±Std':>8} {'Min':>8} {'Max':>8}")
print("-"*60)

cv_results = {}
for nome, r in risultati.items():
    scores = cross_val_score(r['modello'], X_train, y_train, cv=5, scoring='accuracy')
    cv_results[nome] = scores
    print(f"{nome:<25} {scores.mean():>8.3f} {scores.std():>8.3f} {scores.min():>8.3f} {scores.max():>8.3f}")

# Grafico confronto
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Boxplot CV
axes[0].boxplot(cv_results.values(), labels=cv_results.keys(),
               patch_artist=True,
               boxprops=dict(facecolor='#6366f1', alpha=0.7),
               medianprops=dict(color='white', linewidth=2))
axes[0].set_title('Cross-Validation Accuracy (5 fold)', fontweight='bold')
axes[0].set_ylabel('Accuracy')
axes[0].tick_params(axis='x', rotation=10)

# Confronto barplot train vs test
nomi = list(risultati.keys())
acc_train_vals = [risultati[n]['acc_train'] for n in nomi]
acc_test_vals  = [risultati[n]['acc_test']  for n in nomi]
x = np.arange(len(nomi))
w = 0.35
axes[1].bar(x - w/2, acc_train_vals, w, label='Train', color='#6366f1', alpha=0.8)
axes[1].bar(x + w/2, acc_test_vals,  w, label='Test',  color='#10b981', alpha=0.8)
axes[1].set_xticks(x); axes[1].set_xticklabels(nomi, rotation=10)
axes[1].set_title('Train vs Test Accuracy per modello', fontweight='bold')
axes[1].set_ylabel('Accuracy'); axes[1].legend()
axes[1].set_ylim(0.5, 1.05)

plt.tight_layout()
plt.savefig('grafici/confronto_modelli.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ============================================================
# Cella ML5: Analisi dei coefficienti della Logistic Regression
# ============================================================
#
# COME FUNZIONA LA LOGISTIC REGRESSION?
# Il modello calcola una somma pesata delle feature:
#   z = coef[0]*x1 + coef[1]*x2 + ... + coef[n]*xn + intercetta
# Poi applica la funzione sigmoide: probabilità = 1 / (1 + e^-z)
# Se probabilità >= 0.5 → predice High Performance (1)
# Se probabilità <  0.5 → predice Low Performance  (0)
#
# COSA SONO I COEFFICIENTI?
# Ogni coef[i] è il "peso" della feature i-esima.
#   coef > 0: aumentare quella feature aumenta la probabilità
#             di essere classificato come High Performance
#   coef < 0: aumentare quella feature diminuisce la probabilità
#             (la spinge verso Low Performance)
#   |coef| grande: quella feature ha molta influenza
#   |coef| vicino a 0: quella feature conta poco
#
# PERCHÉ USARE IL VALORE ASSOLUTO?
# Sia coef=+2.5 che coef=-2.5 indicano feature molto influenti,
# ma in direzioni opposte. Il valore assoluto misura la
# "forza" dell'influenza indipendentemente dalla direzione.
# ============================================================

lr = risultati['Logistic Regression']['modello']

# Crea un DataFrame con coefficiente e valore assoluto
coef_df = pd.DataFrame({
    'Feature':      feature_cols,
    'Coefficiente': lr.coef_[0],          # valore con segno
    'Importanza':   np.abs(lr.coef_[0])   # valore assoluto = forza dell'effetto
}).sort_values('Importanza', ascending=False)

# Grafico a due pannelli: importanza + direzione dell'effetto
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Pannello 1: importanza (valore assoluto) — come feature importance
colors_imp = ['#4f46e5' if i < 3 else '#a5b4fc' for i in range(len(coef_df))]
bars = axes[0].barh(
    coef_df['Feature'][::-1], coef_df['Importanza'][::-1],
    color=colors_imp[::-1], edgecolor='white'
)
axes[0].set_title('Importanza feature (|coefficiente|)\nLogistic Regression – top 3 in blu scuro',
                  fontweight='bold')
axes[0].set_xlabel('Valore assoluto del coefficiente')
for bar, val in zip(bars, coef_df['Importanza'][::-1]):
    axes[0].text(bar.get_width() + 0.005, bar.get_y() + bar.get_height()/2,
                f'{val:.3f}', va='center', fontsize=9)

# Pannello 2: coefficienti con segno — mostra la DIREZIONE dell'effetto
colors_dir = ['#10b981' if v > 0 else '#ef4444' for v in coef_df['Coefficiente'][::-1]]
axes[1].barh(
    coef_df['Feature'][::-1], coef_df['Coefficiente'][::-1],
    color=colors_dir, edgecolor='white'
)
axes[1].axvline(0, color='white', linewidth=1, linestyle='--')
axes[1].set_title('Direzione dell\'effetto (coefficiente con segno)\nverde = favorisce High Perf. | rosso = favorisce Low Perf.',
                  fontweight='bold')
axes[1].set_xlabel('Coefficiente')

plt.tight_layout()
plt.savefig('grafici/feature_importance_lr.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n=== TOP 5 FEATURE PIÙ INFLUENTI ===")
print(f"{'Feature':<30} {'Coeff.':>8} {'|Coeff.|':>10} {'Direzione':>15}")
print("-"*65)
for _, row in coef_df.head(5).iterrows():
    direzione = "→ High Perf." if row['Coefficiente'] > 0 else "→ Low Perf."
    print(f"  {row['Feature']:<28} {row['Coefficiente']:>8.3f} {row['Importanza']:>10.3f} {direzione:>15}")
```

## Riepilogo finale – Student Exam Performance

### Obiettivo
Predire se uno studente appartiene alla categoria High Performance o Low Performance
basandosi su caratteristiche demografiche e comportamentali,
senza usare il punteggio dell'esame come input.

### Pipeline adottata
1. **EDA**: analisi esplorativa su 6607 studenti originali
2. **Pulizia**: eliminazione 229 righe con NaN (3.5%) → 6378 righe
3. **Preprocessing**:
   - Label Encoding di [N] colonne categoriche
   - Target Variable: Exam_Score >= mediana ([valore]) → 1 (High Performance)
   - Train/Test split: 80% / 20% con stratify
   - Normalizzazione: StandardScaler
4. **Modelli addestrati**: Logistic Regression, Decision Tree, Random Forest

### Perché High/Low Performance e non Promosso/Bocciato?
Con soglia fissa a 60 il dataset risultava sbilanciato (99% High / 1% Low),
rendendo il modello inutile. La mediana garantisce ~50% per classe,
permettendo al modello di imparare pattern reali da entrambe le classi.

### Risultati

| Modello              | Acc. Train | Acc. Test | Gap   | Overfitting | Score corretto |
|----------------------|-----------|-----------|-------|-------------|----------------|
| Logistic Regression  | 0.896     | 0.887     | 0.009 | no ✓        | 0.887          |
| Decision Tree        | 0.845     | 0.824     | 0.021 | [sì/no]     | 0.824          |
| Random Forest        | 1.000     | 0.908     | 0.092 | SI ⚠        | 0.886          |

### Criteri di selezione del modello migliore
Il modello è stato scelto non solo in base all'accuracy sul test set,
ma applicando uno **score corretto** che penalizza l'overfitting:

  score = acc_test − penalità_overfitting

dove la penalità si applica quando il gap train/test supera il 5%.
Questo approccio premia i modelli che generalizzano bene su dati nuovi
piuttosto che quelli che semplicemente memorizzano il training set.

### Modello selezionato: Logistic Regression
- **Accuracy test**: 0.887%
- **Gap train/test**: 0.009% → nessun overfitting rilevato
- **Score corretto**: 0.887
- Pur non avendo l'accuracy assoluta più alta, è il modello più stabile
  e affidabile su dati non visti, il che lo rende preferibile in produzione.

### Analisi delle feature più influenti (coefficienti)
La Logistic Regression non dispone di feature_importances_ come la Random Forest,
ma i suoi **coefficienti** svolgono la stessa funzione: misurano il peso di ogni
variabile nella predizione. Il valore assoluto indica la forza dell'influenza;
il segno indica la direzione (positivo → favorisce High Performance).

| Feature             | Coefficiente | Direzione         |
|---------------------|-------------|-------------------|
| Attendance         | 3.371    | → High Perf. |
| Hours_Studied         | 2.595    | → High Perf. |
| [Previous_Scores         | 1.048    | → High Perf. |

### Conclusioni
- Il modello migliore per robustezza è la **Logistic Regression** (score corretto: [0.887])
- Le variabili più influenti sono: [Attendance], [Hours_Studied], [Previous_Scores]
- [Osservazione principale emersa dall'EDA, es: "gli studenti con più ore di studio
  e alta presenza hanno una probabilità significativamente maggiore di rientrare
  nella categoria High Performance"]
